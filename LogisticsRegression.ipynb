## 1. Label Encoding

```
clm = [1, 4, 7, 10, 11, 15]
for i in clm:
  data[:,i] = le.fit_transform(data[:,i])
  print(np.unique(data[:,i]))
```

## 2. DOB to Age

```
for i in range(data.shape[0]):
  atr = data[i][2].split(' ')
  date_dt2 = datetime.strptime(atr[0], '%Y-%m-%d')
  td = datetime.today()
  diff = td - date_dt2
  age = float(np.around(diff.days/365, 3))
  data[i][2] = age
```

# Models

```
y = y.astype('int')
max_ac = [0]*4
random_state = [0]*4
test_size = [0.1, 0.2, 0.3, 0.4]
for ts in test_size:
  for i in range(120):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = ts, random_state=i)
    
    clf = LR()
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    ac1 = accuracy_score(y_test, y_pred)
    
    if (max_ac[0] < ac1 and ts == 0.1):
      max_ac[0] = ac1
      random_state[0] = i
    if (max_ac[1] < ac1 and ts == 0.2):
      max_ac[1] = ac1
      random_state[1] = i
    if (max_ac[2] < ac1 and ts == 0.3):
      max_ac[2] = ac1
      random_state[2] = i
    if (max_ac[3] < ac1 and ts == 0.4):
      max_ac[3] = ac1
      random_state[3] = i
```

# Normalized Accuracies and Confusion Matrix and Class-wise Accuracy 

```
def checking_accuracy_classwise(y_test,y_pred,class_):
  n=0
  j=0
  i=0
  while (j<y_test.size): #test_accuracy
    if (y_pred[j] == y_test[j] and y_pred[j] == class_):
      n += 1
      i += 1
    elif (y_test[j] == class_):
      i += 1
    j+=1
  accu = n/i
  return accu
```

```
ac_max = 0
split = 0
cm_max = 0
i=0
for ts in test_size:
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = ts, random_state=random_state[i])
  clf = LR()
  clf.fit(X_train, y_train)
  y_pred = clf.predict(X_test)
  ac1 = accuracy_score(y_test, y_pred)
  ac_0 = checking_accuracy_classwise(y_test, y_pred, 0)
  ac_1 = checking_accuracy_classwise(y_test, y_pred, 1)
  cm = confusion_matrix(y_test, y_pred)
  if (ac1>ac_max):
    ac_max = ac1
    split = ts
    cm_max = cm
  print('\033[1m'+"Normalized Accuracy on split of {0:d}/{1:d} is {2:.2f}%".format(int(100-(ts*100)),int(ts*100),ac1*100))
  print('Accuracy for class 0 is {0:.2f}% \nAccuracy for class 1 is {1:.2f}%'.format(ac_0*100,ac_1*100))
  print("Confusion matrix for the same is \n",cm,"\n\n")
  i+=1
```

```
print('\033[1m'+'Maximum normalized accuracy achieved is {2:.2f}% at split of {0:d}/{1:d}'.format(int(100-(split*100)),int(split*100),ac_max*100))
print("Confusion matrix for the same is \n",cm_max)
```
